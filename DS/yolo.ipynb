```python
import json, torch
import cv2
import matplotlib.pyplot as plt

def detect_all(
    image_path,
    imgsz=1280,          # крупнее 640 — лучше ловит длинные/мелкие инструменты
    conf=0.10,           # чуть пониже, чтобы не пропускать слабые детекции
    iou=0.50,            # NMS-порог
    max_det=100,         # с запасом
    tta=False,            # test-time augmentation (flip/scale) повышает recall
    device=None,         # None -> авто, либо 0 для GPU
):
    if device is None:
        device = 0 if torch.cuda.is_available() else 'cpu'

    results = model.predict(
        source=image_path,
        imgsz=imgsz,
        conf=conf,
        iou=iou,
        max_det=max_det,
        augment=tta,
        agnostic_nms=False,
        device=device,
        save=False,       # сами нарисуем ниже
        verbose=False
    )

    r = results[0]
    names = r.names
    dets = []
    for b in r.boxes:
        cls = int(b.cls.item())
        conf = float(b.conf.item())
        x1, y1, x2, y2 = [float(x) for x in b.xyxy.squeeze().tolist()]
        dets.append({
            "class_id": cls,
            "class_name": names[cls],
            "confidence": round(conf, 4),
            "bbox_xyxy": [round(x1, 1), round(y1, 1), round(x2, 1), round(y2, 1)]
        })

    # Нарисовать поверх картинки (как делает .plot()) и показать в ноутбуке
    annotated = r.plot()  # np.ndarray BGR
    annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12, 8))
    plt.imshow(annotated_rgb); plt.axis('off'); plt.title(Path(image_path).name)
    plt.show()

    # JSON-отчёт по всем детекциям
    report = {"image": str(image_path), "detections": dets}
    print(json.dumps(report, ensure_ascii=False, indent=2)[:2000])
    return report, r
```

```python
def yolo_to_xyxy(norm_box, W, H):
    c, x, y, w, h = norm_box  # c - int, остальные float
    x1 = (x - w/2) * W
    y1 = (y - h/2) * H
    x2 = (x + w/2) * W
    y2 = (y + h/2) * H
    return int(x1), int(y1), int(x2), int(y2)

def expand_box(x1,y1,x2,y2, W,H, margin=0.10):
    # Чуть расширим bbox, чтобы целиком захватить инструмент
    w, h = x2-x1, y2-y1
    x1 = max(0, int(x1 - w*margin))
    y1 = max(0, int(y1 - h*margin))
    x2 = min(W, int(x2 + w*margin))
    y2 = min(H, int(y2 + h*margin))
    return x1,y1,x2,y2

def make_crops(split):
    src_img = DATA/'images'/split
    src_lbl = DATA/'labels'/split
    out = CROPS/split
    # создаём папки классов
    for cls in CLASSES:
        (out/cls).mkdir(parents=True, exist_ok=True)
    # обходим изображения
    for img_path in tqdm(list(src_img.glob("*.*")), desc=f"Cropping {split}"):
        img = cv2.imread(str(img_path))
        if img is None: 
            continue
        H, W = img.shape[:2]
        lbl_path = src_lbl/(img_path.stem + ".txt")
        if not lbl_path.exists():
            continue
        with open(lbl_path, "r", encoding="utf-8") as f:
            lines = [l.strip() for l in f if l.strip()]
        for i, line in enumerate(lines):
            parts = line.split()
            cls_id = int(parts[0])
            x,y,w,h = map(float, parts[1:5])
            x1,y1,x2,y2 = yolo_to_xyxy((cls_id,x,y,w,h), W,H)
            x1,y1,x2,y2 = expand_box(x1,y1,x2,y2, W,H, margin=0.12)
            crop = img[y1:y2, x1:x2]
            if crop.size == 0: 
                continue
            # нормируем размер (опционально), но для ImageFolder это не нужно
            out_path = out/CLASSES[cls_id]/f"{img_path.stem}_{i}.jpg"
            cv2.imwrite(str(out_path), crop)
```
